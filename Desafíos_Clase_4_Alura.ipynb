##1° Desafío

$1°$ Promedio del área de todos los inmuebles en los barrios en el dataset. El top 10

```python
# Agrupar por barrio y calcular el área promedio
area_promedio_por_barrio = inmuebles.groupby('Barrio')['Area'].mean()

# Ordenar los barrios por precio promedio y seleccionar los 10 mejores
promedio_areas = inmuebles.groupby('Barrio')['Area'].mean().sort_values(ascending=False)
barrios_valoresAltos = promedio_areas.head(10)

# Calcular el promedio del área promedio para los 10 mejores barrios
promedio_areas = barrios_valoresAltos.mean()

# Imprimir el resultado
print('El promedio del área de los inmuebles en los 10 mejores barrios es: ', promedio_areas)
print('El promedio de los valores altos de los inmuebles en los 10 mejores barrios es: ', barrios_valoresAltos)
barrios_valoresAltos.plot.bar()
```

$$Conclusiones$$
1° El promedio del área de los inmuebles en los 10 mejores barrios es de 762.01 metros cuadrados. Esto indica que en los 10 mejores barrios hay una mayor oferta de inmuebles con áreas más grandes que en otros barrios.$$ $$
2° El barrio con el área promedio más alta es Modelia, con un promedio de 4316.75 metros cuadrados. Esto indica que en Modelia hay una oferta de inmuebles con áreas muy grandes en comparación con los demás barrios del dataset.

$2°$ Consultar otros datos estadísticos, conteo, mediana, valores mínimo y máximo

```python
# Cálculo de los estadísticos
conteo = inmuebles['Area'].count()
mediana = inmuebles['Area'].median()
minimo = inmuebles['Area'].min()
maximo = inmuebles['Area'].max()

# Imprimir los resultados
print('Conteo de observaciones:\n', conteo)
print('Mediana del área:\n', mediana)
print('Valor mínimo del área:\n', minimo)
print('Valor máximo del área:\n', maximo)
```

Según los resultados obtenidos, se puede concluir que el dataset contiene información sobre 9,520 inmuebles con áreas que varían desde 2 hasta 166,243 metros cuadrados. La mediana del área es de 80 metros cuadrados, lo que indica que la mitad de los inmuebles tienen un área menor o igual a 80 metros cuadrados y la otra mitad tiene un área mayor. El valor mínimo del área es de 2 metros cuadrados, lo que sugiere que puede haber algunos inmuebles muy pequeños en el conjunto de datos.

Estos resultados pueden ser útiles para entender la distribución de las áreas de los inmuebles en el conjunto de datos y también pueden ser utilizados como referencia para evaluar la adecuación de un inmueble en términos de su área en relación con el resto de inmuebles del conjunto de datos.

$1°$ Estudiar mejor el histograma de valores, seleccionar 3 tipos de inmuebles (Refinar el gráfico: Títulos, aumentar el tamaño de labels, colores, conclusión de la información)

```python
inmuebles.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

# Seleccionamos los tres tipos de inmuebles más representativos
tipos = ['Apartamento', 'Casa', 'Oficina/Consultorio']
inmuebles_filtrados = inmuebles[inmuebles['Tipo'].isin(tipos)]

# Seleccionamos una paleta de colores adecuada para cada tipo de inmueble
palette = {'Apartamento': '#1f77b4', 'Casa': '#ff7f0e', 'Oficina/Consultorio': '#2ca02c'}

# Graficamos la distribución de los precios por tipo de inmueble
grafica = sns.histplot(data=inmuebles_filtrados, x='Precio_Millon', kde=True, hue='Tipo', palette=palette)

# Añadimos títulos más descriptivos y en negrita
grafica.set(title='Distribución de los valores por tres tipos de inmuebles en Bogotá', xlabel='Precio', ylabel='Frecuencia')

# Aumentamos el tamaño de las etiquetas de los ejes
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Limitamos el eje x para una mejor visualización
plt.xlim((50,1000))

# Añadimos una leyenda más clara
plt.legend(title='Tipo de inmueble', title_fontsize=12, fontsize=10, labels=['Apartamento', 'Casa', 'Oficina/Consultorio'])

# Guardamos la gráfica en formato png
plt.savefig('/content/drive/MyDrive/Alura_course/valor_tipo_inmuebles.png', format='png')

plt.show()
```

$2°$ Precio del m2 por barrio y hacer el gráfico más adecuado para esta nueva variable.

```python
import matplotlib.pyplot as plt
import seaborn as sns

inmuebles['Area'] = inmuebles['Area'].astype(float)
inmuebles['Precio'] = inmuebles['Precio'].astype(float)
precio_m2 = inmuebles['Precio'] / inmuebles['Area']
n_barrios = 10
precio_m2_por_barrio = precio_m2.groupby(inmuebles['Barrio']).mean().sort_values(ascending=False).head(n_barrios)

print('Precio por m2 ', precio_m2_por_barrio)
# Creamos una paleta de colores discreta
colores = plt.get_cmap('tab20').colors[:len(precio_m2_por_barrio)]

# Creamos la gráfica de barras
plt.figure(figsize=(12,6))
plt.bar(precio_m2_por_barrio.index, precio_m2_por_barrio, color=colores)

# Añadimos etiquetas y títulos
plt.xlabel('Barrio', fontsize=14)
plt.ylabel('Precio del metro cuadrado (millones de pesos colombianos)', fontsize=14)
plt.title('Precio del metro cuadrado por barrio en Bogotá', fontsize=16)
plt.xticks(rotation=90, fontsize=12)

plt.savefig('/content/drive/MyDrive/Alura_course/valor_barrios.png', format='png')

plt.show()
```
